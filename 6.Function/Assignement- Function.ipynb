```python
def my_max2(a, b):
    if a > b:
        return a
    return b


def my_max3(a, b, c):
    return my_max2(a, my_max2(b, c))

def my_max4(a, b, c, d):
    return my_max2(a, my_max3(b, c, d))

def my_max5(a, b, c, d, e):
    return my_max2(a, my_max4(b, c, d, e))

def my_max6(a, b, c, d, e, f):
    return my_max2(a, my_max5(b, c, d, e, f))

print(my_max6(5, 3, 8, 2, 10, 3))
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

def nth_prime(n):
    start = 2
    while n > 0:
        if is_prime(start):
            n -= 1
            if n == 0:
                return start
        start += 1
    return -1   # not reachable

for i in range(1, 10):
    print(i, nth_prime(i))
```

```python
def nth_fib(n):
    if n == 1:
        return 0
    if n == 2:
        return 1
    # note: we can merge above 2 lines in single condition. Try
    a, b = 0, 1
    n -= 2
    while n > 0:
        c = a + b
        a = b
        b = c
        n -= 1
    return c

for i in range(1, 10):
    print(i, nth_fib(i))
```
